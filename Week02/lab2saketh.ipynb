## Lab Part1

Please choose appropriate graph and visualize the followings with correct/appropriate labels

* Height only
* Points only
* Height vs Points

```python
import pandas as pd

nba_players = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/all_seasons.csv', index_col = 0)

height = nba_players['player_height']
points = nba_players['pts']
```

```python
import matplotlib.pyplot as plt
```

```python
# Height only
plt.title('Height')
plt.hist(height)
plt.xlabel('Height')
```

```python
#Points only
plt.title('Points')
plt.hist(points)
plt.xlabel('Points')
```

```python
#Height vs Points
plt.scatter(height, points)
plt.xlabel('height')
plt.ylabel('points')
plt.title('height vs points')
plt.draw()
```

## Lab Part2
Please visualize year vs score with correct/appropriate labels

```python
bowling_d = pd.read_csv("https://raw.githubusercontent.com/msaricaumbc/DS_data/master/bowling_stats.csv",
                            header=None,
                            names=['year','city','state','count1','count2'])
def merge_columns(row):
    if pd.isna(row['count2']):
        return row['count1']
    else:
        return row['count1']*1000+row['count2']
    
bowling_d['total']=bowling_d.apply(merge_columns,axis=1)

year = bowling_d['year']
score = bowling_d['total']
```

```python
plt.plot(year, score)
plt.xlabel('year')
plt.ylabel('score')
plt.title('year vs score')
plt.draw()
```
