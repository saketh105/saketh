```python
#extracted mobile number from the notes
```

```python
df['mobile_number']=s.str.extract('([\+]?[(]?[0-9]{3}[)]?[-\s\.]?[0-9]{3}[-\s\.]?[0-9]{4,6})')
df
```

```python
#extracted dates from notes
```

```python
df['dates']=df['notes'].str.findall(r'(\d+[-\/\s]?\d{2}[-\/\s]?\d+)|((\d+)?[-\/\s]?(January?|February?|March?|April?|May?|June?|July?|August?|September?|October?|November?|December?)[-\/\s]?\d+)')
df['dates']  = df['dates'].str[0]
df['dates'] = [item[0] for  item in df['dates']]
df
```

```python
df['end_dates']=df['notes'].str.findall(r'(\d+[-\/\s]?\d{2}[-\/\s]?\d+)|((\d+)?[-\/\s]?(January?|February?|March?|April?|May?|June?|July?|August?|September?|October?|November?|December?)[-\/\s]?\d+)')
df['end_dates']  = df['end_dates'].str[1]
df['end_dates'] = [item[0] for  item in df['end_dates']]
df
```
